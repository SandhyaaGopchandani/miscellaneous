# Preprocessing bitcoin_price.csv to get the date in right format

```python
working_fileDir = os.path.dirname(os.path.realpath('__file__'))
print(working_fileDir)
```

```python
sourceFile = 'bitcoin_price.csv'
source_filePath= os.path.join(working_fileDir, 'data/'+sourceFile)
df = pd.read_csv(source_filePath)

for i, row in df.iterrows():
    dt = datetime.strptime(row['Date'], '%b %d, %Y')        
    dt = dt.strftime('%Y-%m-%d')
    row['Date'] = dt
    df.set_value(i,'Date',dt)

#print(df['Date'])
destFile = 'bitcoin_price_updated.csv'
dest_filePath= os.path.join(working_fileDir, 'stats/'+destFile)
df.to_csv(dest_filePath)
df.head()
```

# Preprocessing bitcoin_dataset.csv to get the date in right format


```python
sourceFile = 'bitcoin_dataset.csv'
source_filePath= os.path.join(working_fileDir, 'data/'+sourceFile)
df = pd.read_csv(source_filePath)

for i, row in df.iterrows():
    dt = datetime.strptime(row['Date'], '%Y-%m-%d 00:00:00')        
    dt = dt.strftime('%Y-%m-%d')
    row['Date'] = dt
    df.set_value(i,'Date',dt)

destFile = 'bitcoin_dataset_updated.csv'
dest_filePath= os.path.join(working_fileDir, 'stats/'+destFile)
df.to_csv(dest_filePath)
df.head()
```

# Day of week analysis

## But just looking at the average might be misleading. Let's look at spread of data per year.

```python
joined_data['year'] = pd.to_datetime(joined_data['Date']).dt.year
joined_data['month'] = pd.to_datetime(joined_data['Date']).dt.month
joined_data['weekday'] = pd.to_datetime(joined_data['Date']).dt.weekday_name

#print(joined_data.weekday)

#week_data=joined_data.groupby(['weekday'])['Close'].mean()

mean_df = joined_data.groupby(['year','weekday'], as_index=False)['Close'].agg({'mean': 'mean'})
std_df = joined_data.groupby(['year','weekday'], as_index=False)['Close'].agg({'std': np.std})
min_df = joined_data.groupby(['year','weekday'], as_index=False)['Close'].agg({'min': np.min})
max_df = joined_data.groupby(['year','weekday'], as_index=False)['Close'].agg({'max': np.max})
median_df = joined_data.groupby(['year','weekday'], as_index=False)['Close'].agg({'median': np.median})



week_data = pd.concat([mean_df, std_df['std'], min_df['min'],max_df['max'], median_df['median']], axis=1)
week_data['var_coeff'] = std_df['std'] / mean_df['mean']

week_data.head(10)
```

### See how prices changes with number of bitcoins mined per day. 

```python
joined_data['abs_btc_count'] = joined_data['btc_total_bitcoins'].diff(-1)
#joined_data['abs_btc_count'] = joined_data['abs_btc_count'].fillna(joined_data['abs_btc_count'].median())
joined_data['abs_btc_count'] = joined_data['abs_btc_count'].fillna(joined_data['abs_btc_count'].mean())

joined_data['abs_btc_count'] =joined_data['abs_btc_count'].abs()
#joined_data['abs_btc_count'] = joined_data['btc_total_bitcoins'] - joined_data['btc_total_bitcoins'].shift(-1)

joined_data_sorted = joined_data.sort_values('abs_btc_count')
joined_data_sorted.head()
fig, ax = plt.subplots(figsize=(12,8))


plt.plot(joined_data_sorted['abs_btc_count'], joined_data_sorted['Close'])
plt.xlabel("BTC count per day", fontsize=15)
plt.ylabel("Bitcoin Price", fontsize=15)
plt.title("Price vs No. of Bitcoins mined per day", fontsize=20)
#plt.savefig('no_ofbtc.png')
plt.show()
```

There is noise in the data but we can see something: when BTC mined per day number is low, price is higher. may be this suggests the economics rule. when supply is lower, demand increases hence increase in price. but if this is the case then price should keep increasing till total of 21M BTC have mined because bitcoin is a fixed supply.


# check price correlation with some other variables in that dataset and see which one are correlated

I chose these variables based on my understanding of bitcoin. To me, these variables might affect the bitcoin price.
- Close: closing price of bitcoin
- btc_market_cap: bitcoin market capitilization for the particular date. Market cap = Total bitcoin volume * price.
- btc_n_transactions_per_block: How many transactions took place per block on a particular date
- abs_btc_count: bitcoin count till date
- btc_hash_rate: Hash rate applied to mine the bitcoins per date
- btc_difficulty: measure of difficulty to mine. The more difficult, the less coins should be mined.
- btc_cost_per_transaction: avg cost of transaction
- btc_n_transactions:

```python
selected_col = joined_data[['Close','btc_market_cap',
                            'btc_avg_block_size',
                            'btc_n_transactions_per_block',
                            'abs_btc_count',
                           'btc_hash_rate',
                           'btc_difficulty',
                           'btc_cost_per_transaction',
                           'btc_n_transactions']]

selected_col.head()
corrmat = selected_col.corr(method='pearson')

columns = ['Close']
my_corrmat = corrmat.copy()
mask = my_corrmat.columns.isin(columns)
my_corrmat.loc[:, ~mask] = 0
#print(my_corrmat)

fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(my_corrmat, annot=False, fmt="f", cmap="Blues") #vmax=1., square=True)
plt.title("Correlation Between Price and other factors", fontsize=15)
#plt.savefig('variablecorrelation.png', bbox_inches='tight')
plt.show()
```

Since, We are concerned with the relationship between bitcoin and other variables, I have made other values 0 for the sake of clarity.

The map shows that BTC price is most correlated with BTC market capitilization, BTC hash rate.

### Correlation between Closing price of bitcoin to Tokyo and US stock closing price


```python
#df[df['first_name'].notnull() & (df['nationality'] == "USA")]


filelist = ["bitcoin_price_updated.csv","afl_stock.csv", "joined_tokyo_stock.csv" ]
header_name = ["bitcoin_price","us_stock_price","tokyo_stock_price"]
temp_list = ["","_snp","_tokyo"]
columns_to_use = []

for index, file_name in enumerate(filelist):
    source_filePath= os.path.join(working_fileDir, 'stats/'+file_name)
    if index == 0:
        df = pd.read_csv(source_filePath, usecols=["Date", "Close"])
        df.columns = ["Date", header_name[index]]
    else:
        temp = pd.read_csv(source_filePath, usecols=["Date", "Close"+temp_list[index]])
        temp.columns = ["Date", header_name[index]]
        df = pd.merge(df, temp, on="Date")
    columns_to_use.append(header_name[index])


cols_to_norm = ['bitcoin_price','us_stock_price','tokyo_stock_price']
df[cols_to_norm] = df[cols_to_norm].apply(lambda x: (x - x.min()) / (x.max() - x.min()))



df['btc_moving_avg'] =  df['bitcoin_price'].rolling(window=10).mean()
df['us_moving_avg'] =  df['us_stock_price'].rolling(window=10).mean()
df['tokyo_moving_avg'] =  df['tokyo_stock_price'].rolling(window=10).mean()



fig, ax = plt.subplots(figsize=(12,8))

x2 = [datetime.datetime.strptime(d,'%Y-%m-%d').date() for d in df.Date]

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())

plt.plot(x2, df.us_moving_avg, label='US stock price')
plt.plot(x2, df.btc_moving_avg, label='bitcoin price')
plt.plot(x2, df.tokyo_moving_avg, label='tokyo stock price')
plt.gcf().autofmt_xdate()
plt.xlabel("Date", fontsize=12)
plt.ylabel("Stock Price", fontsize=12)
plt.title("Stock Price overtime", fontsize=15)
plt.legend()
#plt.savefig('tokyous.png')
plt.show()
```
